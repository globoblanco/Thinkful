```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Plane_Crashes = pd.read_csv('C:\\Users\\gabriel.lobo-blanco\\Downloads\\plane_crashes_data.csv')
```

```python
print(Plane_Crashes)
```

```python
crashes = pd.DataFrame(Plane_Crashes.head(2708))
```

```python
yearly_counts = pd.DataFrame(crashes.year.value_counts())
yearly_counts = yearly_counts.sort_index()
yearly_counts.plot(kind='bar', figsize=(10,5))
plt.ylabel('Crashes per Year')
```

```python
# Split locations on commas
locations = crashes.location.str.split(',').tolist()
locations = [x for x in locations if str(x) != 'nan']
countries = []

# Process countries and clean up text
for entry in locations:
    c = entry[-1]
    c = c.strip()
    countries.append(c)
countries = pd.DataFrame(countries, columns=['country'])
country_counts = pd.DataFrame(countries.country.value_counts().sort_index())
print(country_counts)
```

```python
# Split Operator on commas
operators = crashes.operator.tolist()

companies = pd.DataFrame(operators, columns=['operator'])
operator_counts = pd.DataFrame(companies.operator.value_counts().sort_index())
print(operator_counts)
```

```python
type_counts = pd.DataFrame(crashes.type.value_counts())
type_counts = type_counts.sort_index()
print(type_counts)
```
